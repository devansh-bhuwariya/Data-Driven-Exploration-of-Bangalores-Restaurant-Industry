```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("ZOMATO.xlsx")
```

```python
df.head()
```

```python
df.rename(columns = {'listed_in(city)':'CITY'}, inplace = True)
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
#WE CHECK THE CUSTORMER PREFERENCE ON ONLINE FOOD ORDERING SERVICE ON THE BASIS OF RATING THEY HAVE GIVEN TO HOTELS
#THAT PROVIDE THE FACILITY AND TO THE HOTELS THAT DON'T PROVIDE THE FACILITY
```

```python
plt.figure(figsize = (16,10))
ax = sns.countplot(df['CITY'])
plt.xticks(rotation=90)
#WE CHECK THE CUSTORMER PREFERENCE ON ONLINE FOOD ORDERING SERVICE ON THE BASIS OF RATING THEY HAVE GIVEN TO HOTELS
#THAT PROVIDE THE FACILITY AND TO THE HOTELS THAT DON'T PROVIDE THE FACILITY
```

```python
plt.figure(figsize = (6,6))
sns.countplot(df['online_order'], palette = 'inferno')
#WE CHECK HOW MANY HOTELS PROVIDE THE ONLINE FOOD ORDERINGFACILITY.
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = df)
#WE CHECK THE CUSTORMER PREFERENCE ON ONLINE FOOD ORDERING SERVICE ON THE BASIS OF RATING THEY HAVE GIVEN TO HOTELS
#THAT PROVIDE THE FACILITY AND TO THE HOTELS THAT DON'T PROVIDE THE FACILITY
```

```python
df1 = df.groupby(['CITY','online_order'])['name'].count()
df1.to_csv('CITY_online.csv')
df1 = pd.read_csv('CITY_online.csv')
df1 = pd.pivot_table(df1, values=None, index=['CITY'], columns=['online_order'], fill_value=0, aggfunc=np.sum)
df1
```

```python
df1.plot(kind = 'bar', figsize = (15,8))
#AS WE SEE THAT PEOPLE PREFER TO HAVE ONLINE FOOD ORDERING FACILITY;
#WE CHECK FOR WHICH AREA IS ALREADY DOING WELL IN THAT  ASPECT AND WHICH AREA HAS MORE GROWTH OPPORTUNITY FOR THE SAME.
```

```python
plt.figure(figsize = (6,6))
sns.countplot(df['book_table'], palette = 'rainbow')
#WE CHECK HOW MANY HOTELS PROVIDE THE TABLE BOOKINGFACILITY.
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'book_table', y = 'rate', data = df)
```

```python
df2 = df.groupby(['CITY','book_table'])['name'].count()
df2.to_csv('CITY_booktable.csv')
df2 = pd.read_csv('CITY_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['CITY'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind = 'bar', figsize = (15,8))
#AS WE SEE THAT PEOPLE PREFER TO HAVE TABLE BOOKING FACILITY;
#WE CHECK FOR WHICH AREA HAS MOST HOTELS PROVIDING THE SAME FACILITY 
#AND WHICH AREA DOESN'T HAVE MANY HOTELS PROVIDING THE SAME FACILITY
```

```python
plt.figure(figsize = (14, 8))
sns.boxplot(x = 'Type', y = 'rate', data = df, palette = 'inferno')
#WE CHECK THE CUSTORMER PREFERENCE ON REST_TYPE ON THE BASIS OF RATING THEY HAVE GIVEN TO HOTELS OF THAT REST_TYPE.
#AFTER VISUALIZING WE NOTE THAT PEOPLE PREFER HOTELS WITH BUFFET AND HOTELS THAT HAS DRINKS AND NIGHTLIFE.
```

```python
df4 = df[['CITY', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['CITY'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['votes'])
plt.xticks(rotation = 90)
#WE CHECK FOR AREAS WITH HIGHEST AND LOWEST NUMBER OF CUSTOMERS ON THE BASIS OF VOTES.
#THIS GIVES US A BETTER IDEA ABOUT THE AREA THAT DEMANDS MORE HOTELS.
```

```python
df6 = df[['cuisines', 'votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('votes', ascending=False)
df7.head()
```

```python
df7 = df7.iloc[1:, :]
df7.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df7.index , df7['votes'])
plt.xticks(rotation = 90)
#WE CHECK FOR THE MOST AND LEAST PREFERRED CUSINES THAT CUSTOMERS VOTES FOR.
```
